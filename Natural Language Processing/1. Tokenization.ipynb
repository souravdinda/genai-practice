```python
corpus = """Hello Welcome,to Sourav's world. This is a tokenization practice. 
Let's see how it works!
"""
print(corpus)
```

```python
## Sentence Tokenization
documents = sent_tokenize(corpus)
print(documents)
```

```python
for sentences in sent_tokenize(corpus):
    print(sentences)
```

```python
## Tokenization
## corpus/paragraph to words
from nltk.tokenize import word_tokenize
for words in word_tokenize(corpus):
    print(words)
```

```python
## sentence --> words
for sentences in documents:
    print(word_tokenize(sentences))
```

```python
## Tokenization
## corpus/paragraph to words
## wordpunct_tokenize() --> it will split the words and punctuation separately , i.e. Let's --> ['Let', "'", 's']
from nltk.tokenize import wordpunct_tokenize
for words in wordpunct_tokenize(corpus):
    print(words)
```

```python
## Detokenization
## It is the process of converting a list of tokens back into a string.
from nltk.tokenize import TreebankWordDetokenizer
TreebankWordDetokenizer().detokenize(wordpunct_tokenize(corpus))
```
