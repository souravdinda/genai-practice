```python
'''
Stopword Removal
Stopwords are common words that do not carry much meaning and are often removed from text data to improve the performance of natural language 
processing models. Examples of stopwords include "the", "is", "in", "and", etc. Removing stopwords can help reduce the dimensionality 
of the data and improve the efficiency of text processing tasks. In Python, you can use libraries like NLTK to remove stopwords from your text data. 
Here's an example of how to do this:
Advantages of Stopword Removal:
1. Reduces Noise: Removing stopwords can help reduce noise in the data, making it easier for models to focus on the important words that carry meaning.
2. Improves Efficiency: By reducing the number of words in the text, stopword removal can improve the efficiency of text processing tasks, such as text classification or sentiment analysis.
3. Enhances Model Performance: Removing stopwords can enhance the performance of natural language processing models by allowing them to focus on the more meaningful words in the text data.
Disadvantages of Stopword Removal:
1. Loss of Context: Removing stopwords can sometimes lead to a loss of context, as stopwords can provide important information about the structure and meaning of a sentence.
2. Incomplete Information: In some cases, stopwords may carry important information that is necessary for understanding the meaning of a sentence, and removing them can lead to incomplete information being processed by the model.
3. Language Dependency: The list of stopwords can vary depending on the language being processed, and using an inappropriate list of stopwords can lead to suboptimal results in natural language processing tasks.

'''
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python

paragraph = """I have three visions for India. In 3000 years of our history people from all over the world have come and invaded us, 
captured our lands, conquered our minds. From Alexander onwards the Greeks, the Turks, the Moguls, the Portuguese, the British, the French, 
the Dutch, all of them came and looted us, took over what was ours. Yet we have not done this to any other nation. We have not conquered anyone.
We have not grabbed their land, their culture and their history and tried to enforce our way of life on them. Why? Because we respect the freedom
of others. That is why my FIRST VISION is that of FREEDOM. I believe that India got its first vision of this in 1857, 
when we started the war of Independence. It is this freedom that we must protect and nurture and build on. If we are not free, 
no one will respect us.We have 10 percent growth rate in most areas. Our poverty levels are falling. Our achievements are being globally 
recognised today. Yet we lack the self-confidence to see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect? 
MY SECOND VISION for India is DEVELOPMENT. For fifty years we have been a developing nation. It is time we see ourselves as a developed nation. 
We are among top five nations in the world in terms of GDP. We have a 10 percent growth rate in most sectors. Our poverty levels are falling. 
Our achievements are being globally recognised today. Yet we lack the self-confidence to see ourselves as a developed nation, self-reliant and 
self-assured. Isn’t this incorrect? MY SECOND VISION for India is DEVELOPMENT. For fifty years we have been a developing nation. 
It is time we see ourselves as a developed nation. We are among top five nations in the world in terms of GDP. 
We have nuclear weapons, space technology, our own satellites, our own missile systems. 
What more do we need to be recognised as a developed nation? We have made good progress but we need to accelerate it. 
I have a THIRD VISION. India must stand up to the world. Because I believe that unless India stands up to the world, no one will respect us. 
Only strength respects strength. We must be strong not only as a military power but also as an economic power. Both must go hand-in-hand. 
My good fortune was to have worked with three great minds. Dr.Vikram Sarabhai, of the Dept. of Space, Professor Satish Dhawan, who succeeded him 
and Dr. Brahm Prakash, father of nuclear material. I was lucky to have worked with all three of them closely and 
consider this the great opportunity of my life. I see four milestones in my career"""
```

```python
## stopwords.words('english') will give us the list of stopwords in English language. We can use this list to remove stopwords from our paragraph.
stopwords.words('english')
```

```python
stopwords.words('German')
```

```python
import nltk
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
## sent_tokenize() function is used to split the paragraph into sentences.
sentences = nltk.sent_tokenize(paragraph)
sentences
```

```python
## Apply stopword removal and Filter and then Apply stemming to each sentence in the paragraph
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) ## convert the list of words back to a sentence
sentences
```

```python
## Apply stopword removal and Filter and then Apply Snowball to each sentence in the paragraph
sentences = nltk.sent_tokenize(paragraph)
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [snowball_stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) ## convert the list of words back to a sentence
sentences
```

```python
## Apply stopword removal and Filter and then Apply Lemmatization to each sentence in the paragraph
sentences = nltk.sent_tokenize(paragraph)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) ## convert the list of words back to a sentence
sentences

## Why lowercase in lemmatization? Because lemmatization is case sensitive and it will not work properly if the words are in uppercase.
```
