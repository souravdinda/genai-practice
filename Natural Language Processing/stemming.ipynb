# Stemming
Stemming is the process of reducing a word to its base or root form. The root form may not be a valid word in the language, but it is used to group together different forms of the same word. For example, the words "running", "runner", and "ran" can all be reduced to the root form "run". Stemming is commonly used in natural language processing tasks such as information retrieval, text mining, and sentiment analysis to improve the performance of algorithms by reducing the number of unique words in a text corpus.

Reduces vocabulary size — groups word forms (running, runs, ran → run) so models see fewer unique tokens.

Improves recall in search/IR — queries match more relevant documents with different word forms.

Reduces sparsity — benefits statistical models and embeddings when training data is limited.

Speeds downstream processing and model training (smaller feature set).

Simplifies feature engineering for classification/clustering.

Tradeoff: can hurt precision (over-/under-stemming) and produce nonwords; use lemmatization if you need valid dictionary forms.


```python
words = ["eating", "eaten", "eat", "eats","runner", "running", "runs", "ran",
         "program", "programmer", "programming", "programs","swim", "swimming",
         "swims", "swam","write", "writer", "writing", "writes","history", "historical", "historian", "histories"]
```

```python
## Porter Stemmer
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
for word in words:
    print(f"for {word}: {porter_stemmer.stem(word)}")

```

```python
## history, historical, historian, histories --> histori.
##Sometimes it may not give the correct root word. It is not a real word. 
# It is just a stemmed word.
porter_stemmer.stem("congratulations")
```

```python
porter_stemmer.stem("sitting")
```


Regexp Stemmer class
RegexpStemmer allows you to define custom stemming rules using regular expressions to remove suffixes from words.

How it works:

You specify a regex pattern to match suffixes you want to strip.
Words matching the pattern get those suffixes removed.
More flexible than PorterStemmer—you control the rules.


```python
from nltk.stem import RegexpStemmer
## It will remove the specified suffix from the word. 
# It is a simple stemmer. ing may be the suffix. 
# It will remove ing from the word. It is not a real word. It is just a stemmed word.
## min is the minimum length of the word after stemming.
# If the length of the word after stemming is less than min, then it will not stem the word.
## $ is used to specify the suffix. It means that the suffix should be at the end of the word.
regexp_stemmer = RegexpStemmer("ing$|s$|e$|able$", min=4)
```

```python
regexp_stemmer.stem("eating")
```

```python
regexp_stemmer2 = RegexpStemmer("ing|s$|e$|able$", min=4)
regexp_stemmer2.stem("ingeating")
```

```python
## Snowball Stemmer
## It is an improved version of the Porter Stemmer. 
# It is more aggressive than the Porter Stemmer.
# It is also known as the Porter2 Stemmer. 
# It is available in 15 languages.
# It is more accurate than the Porter Stemmer.

from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer("english")
for word in words:
    print(f"for {word}: {snowball_stemmer.stem(word)}")
```

```python
porter_stemmer.stem("fairly"), snowball_stemmer.stem("fairly")
```

```python
porter_stemmer.stem("sportingly"), snowball_stemmer.stem("sportingly")
```

```python
## Disadvantages of Stemming
porter_stemmer.stem("goes"), snowball_stemmer.stem("goes")
```

```python
## Disadvantages of Stemming
# 1. It may not give the correct root word. It is not a real word. It is just a stemmed word.
# 2. It may not be able to handle irregular words. For example, 
# the stem of "ran" is "ran" and not "run".
# 3. It may not be able to handle words with multiple meanings.
# For example, the stem of "bark" is "bark" and not "bark" (the- sound made by a dog).
```

```python
## Lemmatization
## It is the process of reducing a word to its base form (lemma).
## It is more accurate than stemming.
## It is based on the WordNet lexical database.
'''
Lemmatization is a more sophisticated text normalization technique compared to stemming, 
offering several key advantages. Unlike stemming, which blindly removes suffixes
and may produce non-words (e.g., "history" → "histori"), 
lemmatization uses a lexical database like WordNet to convert words to their valid dictionary
forms—their lemmas. This ensures that "running" becomes "run", "went" becomes "go", and "better" becomes "good", all of which are real, meaningful words. 
This accuracy is particularly valuable in applications requiring semantic understanding, such as sentiment analysis, machine translation, and question-answering systems, where preserving word meaning is critical. 
Additionally, lemmatization handles irregular words better than stemming; 
it can correctly identify that "ran" is a form of "go", not just keep it as "ran". 
By reducing vocabulary to canonical dictionary forms, 
lemmatization improves model performance on smaller datasets and enables more meaningful 
feature representations. However, the trade-off is computational cost—lemmatization is slower than stemming because it requires morphological analysis and dictionary lookups. It's best used when accuracy and semantic preservation matter more than speed, 
making it ideal for NLP tasks where understanding the true meaning of words is essential.
'''
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("going",pos="v") 

## pos is the part of speech. It is used to specify the part of speech of the word. 
# It is optional. The default value is "n" (noun).
## pos can be "n" (noun), "v" (verb), "a" (adjective), "r" (adverb), "s" (satellite adjective).
```

```python
for word in words:
    print(f"for {word}: {lemmatizer.lemmatize(word)}")
```

```python
lemmatizer.lemmatize("goes",pos="v"), lemmatizer.lemmatize("fairly",pos="r"), lemmatizer.lemmatize("sportingly")    
```
