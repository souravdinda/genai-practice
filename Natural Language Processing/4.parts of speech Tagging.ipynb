```python
## Parts of Speech Tagging
## POS tagging is the process of assigning a part of speech to each word in a sentence. It is used to identify the grammatical structure of a sentence and to understand the meaning of the sentence.
## POS tagging is done using the pos_tag() function in NLTK. It takes a list of words as input and returns a list of tuples, where each tuple contains a word and its corresponding part of speech tag.
## The pos_tag() function uses the Penn Treebank POS tags.
## For example, the word 'run' can be a verb or a noun. The pos_tag() function will assign the tag 'VB' to 'run' when it is used as a verb and 'NN' when it is used as a noun.
## The pos_tag() function uses the default tagger, which is a rule-based tagger. It uses a set of rules to assign tags to words based on their context. The default tagger is not very accurate and can make mistakes. For example, it may assign the tag 'NN' to the word 'run' when it is used as a verb.
## To improve the accuracy of POS tagging, we can use a trained tagger. NLTK provides several trained taggers, such as the UnigramTagger, BigramTagger, and TrigramTagger.
## The UnigramTagger assigns tags to words based on their frequency in the training data. The BigramTagger assigns tags to words based on the previous word and the current word. The TrigramTagger assigns tags to words based on the previous two words and the current word.
```

```python
## find pos tags for paragraph
import nltk
nltk.download('averaged_perceptron_tagger_eng')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
sentences = nltk.sent_tokenize(paragraph)
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [ word for word in words if word not in set(stopwords.words('english'))]
    tagged = nltk.pos_tag(words)
    print(tagged)

### PRP -> Personal Pronoun
## VBP -> Verb, non-3rd person singular present
## CD -> Cardinal Number
## NNS -> Noun, plural
## IN -> Preposition or subordinating conjunction
## DT -> Determiner
## NN -> Noun, singular or mass
## VBD -> Verb, past tense
## VBN -> Verb, past participle
## JJ -> Adjective
## NNP -> Proper noun, singular
## NNPS -> Proper noun, plural
    
```
